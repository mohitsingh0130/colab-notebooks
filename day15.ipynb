```
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
```

```
from sklearn.datasets import make_circles
n_samples=500
features,labels=make_circles(noise=0.1,factor=0.2,shuffle=True,n_samples=500,random_state=123)
plt.scatter(features[:,0],features[:,1],c=labels)
plt.show()
```

```
labels=to_categorical(labels,2)
```

```
model= Sequential()
model.add(Dense(units=3,input_dim=2,activation='sigmoid'))
model.add(Dense(units=2,activation='sigmoid'))
model.compile(Adam(lr=0.01),loss='categorical_crossentropy',metrics=['accuracy'])
```

```
print(model.summary())
```

```
h=model.fit(features,labels,epochs=100,verbose=1)
```

```
plt.plot(h.history['acc'],label='acc')
plt.plot(h.history['loss'],label='loss')
plt.legend()
plt.show()
```

```
x_max,x_min=features[:,0].min()-1,features[:,0].max()+1
y_max,y_min=features[:,1].min()-1,features[:,1].max()+1
xx,yy=np.meshgrid(np.linspace(x_min,x_max),np.linspace(y_min,y_max))
grid=np.c_[xx.ravel(),yy.ravel()]
pred=model.predict(grid)
pred=pred.reshape(xx.shape)
```

```
plt.contourf(xx,yy,pred)
plt.scatter(features[:n_samples,0],features[:n_samples,1])
plt.scatter(features[n_samples:,0],features[n_samples:,1])
plt.show()
```
